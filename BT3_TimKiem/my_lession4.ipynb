#### 1. Biễu diễn bằng danh sách / ma trận kề

````python
import networkx as nx
import matplotlib.pyplot as plt

# Danh sách kề
ADJ = {
    'A': {'C', 'E', 'D', 'F'},
    'B': set(),
    'C': {'H'},
    'D': {'H', 'E'},
    'E': {'K', 'I'},
    'F': {'I', 'G'},
    'G': set(),
    'H': {'K'},
    'I': {'B', 'K'},
    'K': {'B'}
}

# Tạo đồ thị có hướng
G = nx.DiGraph()
for node, neighbors in ADJ.items():
    for neighbor in neighbors:
        G.add_edge(node, neighbor)

# Vẽ đồ thị
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Tạo vị trí các node
nx.draw(G, pos, with_labels=True, node_size=2000, node_color='lightblue', edge_color='gray', font_size=12, font_weight='bold')
plt.title("Biểu diễn đồ thị từ danh sách kề")
plt.show()
````

````python
def LuuDoThi(G, file_path, verbose=True):
    import os

    # Tạo thư mục chứa file_path nếu chưa tồn tại
    file_dir = os.path.dirname(file_path)
    if file_dir and not os.path.exists(file_dir):
        os.makedirs(file_dir)
        if verbose:
            print(f"+ Tạo thư mục: {file_dir}")

    # Lưu đồ thị xuống tập tin
    with open(file_path, "wt") as file:
        for node in G:
            neighbors = " ".join(G[node])  # Lấy danh sách đỉnh kề
            file.write(f"{node}: {neighbors}\n")

        if verbose:
            print(f"Lưu đồ thị thành công xuống tập tin: {file_path}")

# Thử nghiệm
ADJ = {
    'A': {'C', 'E', 'D', 'F'},
    'B': set(),
    'C': {'H'},
    'D': {'H', 'E'},
    'E': {'K', 'I'},
    'F': {'I', 'G'},
    'G': set(),
    'H': {'K'},
    'I': {'B', 'K'},
    'K': {'B'}
}

LuuDoThi(ADJ, "dske1.txt", verbose=True)

# Đọc lại tập tin để kiểm tra
with open("dske1.txt", "rt") as file:
    lines = file.readlines()
    for line in lines:
        print(line, end="")
````

````python
import os
import pprint

def DocDoThi(file_path, verbose=True):
    """
    Đọc danh sách kề từ tập tin và trả về đồ thị dưới dạng dictionary.
    
    return:
    + None: nếu đọc thất bại
    + Dictionary chứa đồ thị nếu đọc thành công
    """
    if not os.path.exists(file_path):
        if verbose:
            print(f"Không tìm thấy tập tin: {file_path}")
        return None

    G = {}
    with open(file_path, "rt") as file:
        for line in file:
            parts = line.strip().split(":")
            if len(parts) == 2:
                node = parts[0].strip()
                neighbors = set(parts[1].split()) if parts[1] else set()
                G[node] = neighbors
            elif len(parts) == 1 and parts[0]:  # Trường hợp đỉnh không có cạnh nào
                G[parts[0].strip()] = set()

    if verbose:
        print(f"Đọc đồ thị thành công từ tập tin: {file_path}")
    
    return G

# Thử nghiệm đọc lại tập tin đã lưu
G1 = DocDoThi("dske1.txt", verbose=True)
pprint.pprint(G1)
````

### Bài 2. Tìm kiếm đường đi trên đồ thị

#### 1. Tìm kiếm theo chiều rộng (BFS)

````python
import pprint

def BFS(G, start, goal):
    """
    Duyệt đồ thị theo chiều rộng (BFS) để tìm đường đi từ start đến goal.

    return:
    + Danh sách chứa đường đi từ start đến goal.
    + [] nếu không tìm thấy đường đi.
    + None nếu start hoặc goal không hợp lệ.
    """
    if start not in G or goal not in G:
        return None  # Đỉnh không hợp lệ
    
    path = {}  # Lưu vết đường đi
    s_open = [start]  # Danh sách đỉnh cần duyệt (hàng đợi BFS)
    s_closed = set()  # Tập đỉnh đã duyệt
    
    path[start] = None  # Đỉnh xuất phát không có cha
    
    while s_open:
        current = s_open.pop(0)  # Lấy đỉnh đầu hàng đợi
        s_closed.add(current)  # Đánh dấu đã duyệt
        
        if current == goal:  # Nếu tìm thấy đích, kết thúc tìm kiếm
            break
        
        for neighbor in G[current]:
            if neighbor not in s_closed and neighbor not in s_open:
                s_open.append(neighbor)  # Thêm vào hàng đợi
                path[neighbor] = current  # Ghi lại đường đi

    # Truy vết đường đi từ goal về start
    if goal not in path:
        return []  # Không có đường đi

    route = []
    step = goal
    while step is not None:
        route.append(step)
        step = path[step]
    
    return route[::-1]  # Đảo ngược để có đường đi đúng thứ tự


# Ví dụ đồ thị
G = {
    'A': {'C', 'E', 'D', 'F'},
    'B': {},
    'C': {'H'},
    'D': {'H', 'E'},
    'E': {'K', 'I'},
    'F': {'I', 'G'},
    'G': {},
    'H': {'K'},
    'I': {'B', 'K'},
    'K': {'B'}
}

# Kiểm tra BFS từ A đến B
result = BFS(G, 'A', 'B')
print("Đường đi từ A đến B:", result)
````

````python
def find_path(path, start, goal):
    """
    Cho mảng truy hồi đường, tìm danh sách đỉnh từ start --> goal.
    """
    if goal not in path:
        return []  # Không có đường đi

    result = []
    step = goal

    while step is not None:
        result.append(step)
        step = path.get(step)  # Truy vết ngược lại

    return result[::-1]  # Đảo ngược danh sách để có thứ tự từ start -> goal
````

#### 2. Tìm kiếm theo chiều rộng (DFS)

````python
def DFS(G, start, goal):
    """
    Duyệt đồ thị theo chiều sâu (DFS) để tìm đường đi từ start đến goal.

    return:
    + Danh sách chứa đường đi từ start đến goal.
    + [] nếu không tìm thấy đường đi.
    + None nếu start hoặc goal không hợp lệ.
    """
    if start not in G or goal not in G:
        return None  # Đỉnh không hợp lệ
    
    path = {}  # Lưu vết đường đi
    s_open = [start]  # Ngăn xếp DFS
    s_closed = set()  # Tập đỉnh đã duyệt
    
    path[start] = None  # Đỉnh xuất phát không có cha
    
    while s_open:
        current = s_open.pop()  # Lấy đỉnh trên cùng của stack
        s_closed.add(current)  # Đánh dấu đã duyệt
        
        if current == goal:  # Nếu tìm thấy đích, kết thúc tìm kiếm
            break
        
        for neighbor in G[current]:
            if neighbor not in s_closed and neighbor not in s_open:
                s_open.append(neighbor)  # Thêm vào stack
                path[neighbor] = current  # Ghi lại đường đi

    # Truy vết đường đi từ goal về start
    if goal not in path:
        return []  # Không có đường đi

    route = []
    step = goal
    while step is not None:
        route.append(step)
        step = path[step]
    
    return route[::-1]  # Đảo ngược để có đường đi đúng thứ tự
````
